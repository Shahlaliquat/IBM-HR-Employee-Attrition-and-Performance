# Data Loading and Preprocessing

```python
import pandas as pd
```

```python
#load the dataset
df=pd.read_csv("C:\\Users\\PMLS\\Downloads\\Data analytics\\Portfolio projects\\IBM_HR_Analytics.csv")
# Display the first few rows to see data
print(df.head())
```

```python
#Checking for missing values 
print(df.isnull().sum())
```

```python
# Calculate the overall attrition rate
attrition_rate = df['Attrition'].value_counts(normalize=True)['Yes'] * 100

print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
#Encode Categorical variables
df['Attrition']=df['Attrition'].map({'Yes': 1 ,"No":0 })
print(df['Attrition'])
```

# Statistical Analysis

```python
# Group by Attrition and calculate mean for each group
attrition_grouped = df.groupby('Attrition').mean()

# Boxplot for MonthlyIncome by Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Attrition')
plt.show()
```

# Hypothesis Testing

```python
from scipy.stats import ttest_ind

# T-test for MonthlyIncome
left = df[df['Attrition'] == 1]['MonthlyIncome']
stayed = df[df['Attrition'] == 0]['MonthlyIncome']
t_stat, p_val = ttest_ind(left, stayed)
print(f'T-test results: t-statistic={t_stat}, p-value={p_val}')
```

# Predictive Modeling (Linear Regression)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define the features and target
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the model
lr_model = LinearRegression()

# Train the model
lr_model.fit(X_train, y_train)

# Make predictions
y_pred = lr_model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
